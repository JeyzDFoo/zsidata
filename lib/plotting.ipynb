```python
np.random.seed(200)  # seed the random number generator, makes random numbers static with each run
data = {'a': np.arange(5), #0, 1, 2, 3, 4
        'c': np.random.randint(0, 100, 5), #(low, high, number)
        'd': np.random.randn(5) #number of random numbers between 0 and 1
       }

data['b'] = data['a'] + 10 * np.random.randn(5)
data['d'] = np.abs(data['d']) * 100 #takes absolute value of array d and multiplies by 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data=data) #(x-axis data, y-axis data, 
ax.set_xlabel('Data A')
ax.set_ylabel('Data B')
ax.set_title('Example Scatter Plot');
print(data['a'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x,y)


X0 = -2
X1 = +2

#electric field lines
F = (X-X0)/np.sqrt((X-X0)**2+Y**2) - (X-X1)/np.sqrt((X-X1)**2+Y**2) #defining a function

#equipotential lines
E = -abs(1/np.sqrt((X-X0)**2 + Y**2) - 1/np.sqrt((X-X1)**2 + Y**2))


plt.contour(X, Y, F, 10, colors='black') #(x, y, function, number of lines, color)
plt.contourf(X, Y, F, 100, cmap='rainbow') #contour faces with color map

#plot equipotential lines (dashes)
plt.contour(X, Y, E, levels=[-0.5, -0.4, -0.3, -0.2, -0.1, -0.05], colors='black') #(x, y, function, number of lines, color)

plt.savefig('example figure')

```
